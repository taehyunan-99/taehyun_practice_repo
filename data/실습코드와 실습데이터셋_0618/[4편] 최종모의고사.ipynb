### 1. Auto MPG 데이터셋(df)에서 df1과 df2를 다음과 같이 생성한 후, model_year 컬럼을 기준으로 내부 조인(inner join)하여 df_merge에 저장하세요.
 - df1 : "model_year", "mpg", "cylinders", "displacement" 컬럼만 포함

# 유형6. 이상치 제거

### 1. 연비(MPG)가 50을 초과하는 이상치를 제거하세요.
 - mpg 값이 50을 초과하는 행을 제거하세요.
 - 제거한 데이터는 df로 재정의하세요.

```python
df = df.drop(df[df.mpg > 50].index, axis=0)
df
```

### 답 작성 셀

### 2. 마력(Horsepower)이 250을 초과하는 이상치를 제거하세요.
 - horsepower 값이 250을 초과하는 행을 제거하세요.
 - 제거한 데이터는 df로 재정의하세요.

```python
df = df.drop(df[df.horsepower > 250].index, axis=0)
df
```

### 답 작성 셀

### 3. 가속력(Acceleration)이 30을 초과하는 이상치를 제거하세요.
 - acceleration 값이 30을 초과하는 행을 제거하세요.
 - 제거한 데이터는 df로 재정의하세요.

```python
df = df.drop(df[df.acceleration > 30].index, axis=0)
df
```

### 답 작성 셀

### 4. 마력(Horsepower)의 이상치를 IQR(사분위 범위) 방법으로 제거하세요.
 - IQR(Interquartile Range)을 활용하여 horsepower의 이상치를 제거하세요.
 - 이상치 기준 : Q1 - 1.5 * IQR, Q3 + 1.5 * IQR
 - 제거된 데이터는 df로 재정의하세요.

```python
Q1 = df["horsepower"].quantile(0.25)
Q3 = df["horsepower"].quantile(0.75)
IQR = Q3 - Q1

df = df[(df["horsepower"] >= Q1 - 1.5 * IQR) & (df["horsepower"] <= Q3 + 1.5 * IQR)]
df
```

### 답 작성 셀

### 5. 가속력(Acceleration)의 이상치를 표준편차(Z-score) 방법으로 제거하세요.
 - Z-score(표준화 점수)를 활용하여 acceleration의 이상치를 제거하세요.
 - 이상치 기준 : 절댓값이 3을 초과하는 값 제거

```python
from scipy import stats

df[(abs(stats.zscore(df["acceleration"])) > 3)]
```

### 6. matplotlib을 사용하여 훈련 데이터 손실 값(Training Loss)과 검증 데이터 손실 값(Validation Loss)의 차이를 시각화하세요.
- 훈련 손실 값과 검증 손실 값의 차이를 계산하여 그래프로 표현하세요.

```python
import matplotlib.pyplot as plt
import numpy as np

# 모델 학습 과정에서 저장된 손실(loss) 값 가져오기
history = model.history.history  

# 훈련 손실과 검증 손실 차이 계산
loss_diff = np.array(history['loss']) - np.array(history['val_loss'])
epochs = range(len(loss_diff))  # 전체 epoch 개수

# 그래프 그리기
plt.plot(epochs, loss_diff, label='Training Loss - Validation Loss', marker='o', linestyle='dashed')
plt.xlabel('Epochs')  # X축: Epoch (학습 반복 횟수)
plt.ylabel('MSE Difference')  # Y축: 손실 값 차이
plt.title('Training Loss - Validation Loss Difference')  # 그래프 제목
plt.axhline(0, color='red', linestyle='dotted')  # 기준선 추가
plt.legend()
plt.grid(True)
plt.show()
```
